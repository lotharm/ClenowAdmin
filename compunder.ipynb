```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def generate_random_series(length=100):
    # Generiere ein Array mit Zufallszahlen im Intervall [0, 1]
    random_numbers = np.random.rand(length)
    
    # Erstelle eine Pandas Series aus dem Array
    random_series = pd.Series(random_numbers)

    return random_series



def generate_booleanSeries(I_series, pwin=0.5):
    # Erstelle eine leere Serie R mit denselben Indizes wie I
    R_series = pd.Series(index=I_series.index)
    
    # Setze den Wert in R auf 1, wenn I größer als 100 ist, sonst auf 0
    R_series[I_series >= pwin] = 0
    R_series[I_series < pwin] = 1
    
    return R_series

def generate_PnLSeries(I_series,sw,sl,Bdgt):
    # Erstelle eine leere Serie R mit denselben Indizes wie I
    R_series = pd.Series(index=I_series.index)
    R_series[0]=Bdgt
    # Setze den Wert in R auf 1, wenn I größer als 100 ist, sonst auf 0

    
    for r in range(1,len(R_series)):
        R_series[r] = (1+(sw*I_series[r]+(I_series[r]-1)*sl))*R_series[r-1]
        #print(I_series[r],R_series[r-1],(1+(sw*I_series[r]+(I_series[r]-1)*sl))*R_series[r-1])
    #print(R_series[:20])
    return R_series


def hist_cum_plot(df,title="Histogramn and Cum Dist.",bin=30,blocker=False): 
    fig, ax = plt.subplots(2,figsize=(8,6))
    fig.suptitle(title,fontsize = 6)
    sns.histplot(df,bins=bin,stat="density",ax=ax[0])
    
    print("------------------------------------------")
    print(title+"\n")
    print(df.describe())
    print("------------------------------------------")
    sns.histplot(df,bins=bin,cumulative=1,stat="density",ax=ax[1])
    plt.show(block=blocker)

def hist_cum_plot_simple(df,title="Histogramn and Cum Dist.",bin=30,blocker=False): 
    fig, ax = plt.subplots(1,figsize=(6,4))
    fig.suptitle(title,fontsize = 8)
    sns.histplot(df,bins=bin,stat="density",ax=ax)
    print(df.describe())
    plt.show(block=blocker)

```

```python

Risk=0.008
RiskMultiple= 2

PWinner=4/9
PLooser=1-PWinner
Budget = 100

NumIters=10

SizeWinner=Risk*RiskMultiple
SizeLooser=Risk
# Beispielhafte Verwendung:
```

```python
PnL_frame=pd.DataFrame()


for i in range(500):
    random_series = generate_random_series()
    boolean_series = generate_booleanSeries(random_series, PWinner)
    PnL_series = generate_PnLSeries(boolean_series,SizeWinner,SizeLooser,Budget)
    
    PnL_frame = pd.concat([PnL_frame,PnL_series],axis=1)
```

```python
PnL_frame["Mean"]=PnL_frame.mean(axis=1)
PnL_frame["Mean"].plot()
```

```python

PnL_frame["Min"]=PnL_frame.min(axis=1)
PnL_frame["Max"]=PnL_frame.max(axis=1)
PnL_frame[["Mean","Min"]].plot() 
```

```python
hist_cum_plot(PnL_frame.iloc[-1],title="Histogramn and Cum Dist.",bin=50,blocker=False)
```

```python
hist_cum_plot_simple(PnL_frame.iloc[-1],title="Histogramn and Cum Dist.",bin=50,blocker=False)
```
